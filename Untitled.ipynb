```python
import random

# Word bank
WORD_LIST = ["banana", "python", "developer", "algorithm", "keyboard", "internet", "hangman"]

MAX_TRIES = 6

def choose_word():
    return random.choice(WORD_LIST)

def display_word(word, guessed_letters):
    return ' '.join([letter if letter in guessed_letters else '_' for letter in word])

def get_guess(used_letters):
    while True:
        guess = input("Enter a letter: ").lower()
        if len(guess) != 1:
            print("Please enter just one letter.")
        elif not guess.isalpha():
            print("Only letters are allowed.")
        elif guess in used_letters:
            print("You've already guessed that letter.")
        else:
            return guess

def draw_hangman(tries_left):
    stages = [
        """
         -----
         |   |
         O   |
        /|\\  |
        / \\  |
             |
        =======""",
        """
         -----
         |   |
         O   |
        /|\\  |
        /    |
             |
        =======""",
        """
         -----
         |   |
         O   |
        /|\\  |
             |
             |
        =======""",
        """
         -----
         |   |
         O   |
        /|   |
             |
             |
        =======""",
        """
         -----
         |   |
         O   |
         |   |
             |
             |
        =======""",
        """
         -----
         |   |
         O   |
             |
             |
             |
        =======""",
        """
         -----
         |   |
             |
             |
             |
             |
        ======="""
    ]
    print(stages[MAX_TRIES - tries_left])

def play_game():
    word = choose_word()
    guessed_letters = set()
    used_letters = set()
    tries = MAX_TRIES

    print("🎮 Welcome to Hangman!\n")
    
    while tries > 0:
        draw_hangman(tries)
        print("Word:", display_word(word, guessed_letters))
        print("Guessed letters:", ' '.join(sorted(used_letters)))
        
        guess = get_guess(used_letters)
        used_letters.add(guess)

        if guess in word:
            guessed_letters.add(guess)
            if all(letter in guessed_letters for letter in word):
                print("\n🎉 You guessed the word:", word)
                print("✅ You win!")
                break
        else:
            print("Wrong guess!")
            tries -= 1

    else:
        draw_hangman(0)
        print("\n💀 Game over! The word was:", word)

def main():
    while True:
        play_game()
        again = input("\nPlay again? (y/n): ").lower()
        if again != 'y':
            print("Thanks for playing Hangman!")
            break

if __name__ == "__main__":
    main()
```
